```python
import pandas as pd
import numpy as np
from surprise import KNNBasic,Reader, Dataset, SVD
from surprise.model_selection import KFold, cross_validate
from surprise import Dataset, BaselineOnly, Reader
from surprise import SVD, KNNBasic, KNNWithMeans
from surprise.model_selection import cross_validate
import matplotlib.pyplot as plt
import seaborn as sns
from plotly.offline import init_notebook_mode, plot, iplot
import plotly.graph_objs as go
from IPython.display import Image
init_notebook_mode(connected=True)
%matplotlib inline
```

```python
DataPath = "/content/ratings_small.csv"


reader = Reader(line_format="user item rating timestamp", sep=',', skip_lines=1)

data = Dataset.load_from_file(DataPath, reader=reader)
```

```python
pmf_svd = SVD()
cv_pmf = cross_validate(pmf_svd, data, measures=['RMSE', 'MAE'], cv=5, verbose=True);
```

```python
def CrossValidate(algorithms, cv=5):
	means = {}
	for name in algorithms:
		output = cross_validate(algorithms[name], data, measures=['RMSE', 'MAE'], cv=5, verbose=False)
		result = pd.DataFrame.from_dict(output)
		means[name] = result.mean()
	bm = pd.DataFrame.from_dict(means, orient='index')
	bm.drop(columns=['fit_time', 'test_time'], inplace=True)
	return bm
```

```python
algorithm_ub = {
	'PMF': SVD(biased=False),
	'UserBased': KNNBasic(sim_options={'user_based': True}),
}

bms = CrossValidate(algorithm_ub)
     
```

```python
similarities = ['cosine', 'msd', 'pearson']
measures = ['RMSE', 'MAE']
results = {
    'ubcf': dict(),
    'ibcf': dict()
}

for sim in similarities:
  usercf = KNNBasic(sim_options={'name': sim, 'user_based': True})
  itemcf = KNNBasic(sim_options={'name': sim, 'user_based': False})

  print(f"Computing {sim} for User Based CF")
  cross_validate_ubcf = cross_validate(usercf, data, measures, verbose=True, cv=5)

  print(f"Computing {sim} for Item Based CF")
  cross_validate_ibcf = cross_validate(itemcf, data, measures, verbose=True, cv=5)

  results['ubcf'][sim] = cross_validate_ubcf
  results['ibcf'][sim] = cross_validate_ibcf
```

```python
ptdata = {
    'mae': {
      'cosine': [],
      'msd': [],
      'pearson': []
    },
    'rmse': {
      'cosine': [],
      'msd': [],
      'pearson': []
    }
}

for k, v in results.items():
  for key, result in results[k].items():
    ptdata['mae'][key].append(result['test_mae'].mean())
    ptdata['rmse'][key].append(result['test_rmse'].mean())
```

```python
labels = ['User based Collaborative Filtering', 'Item based Collaborative Filtering']

x = np.arange(len(labels))
width = 0.1

plt.bar(x-0.1, ptdata['mae']['cosine'], width, color='blue')
plt.bar(x, ptdata['mae']['msd'], width, color='red')
plt.bar(x+0.1, ptdata['mae']['pearson'], width, color='green')

plt.xticks(x, labels)
plt.ylabel("MAE")
plt.yticks()
plt.legend(["Cosine", "MSD", "Pearson"])
plt.tight_layout()

plt.show()
```
